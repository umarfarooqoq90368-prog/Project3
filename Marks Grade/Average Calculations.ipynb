```python

subjects = int(input("Enter number of subjects: "))
total = 0
for i in range(subjects):
    marks = float(input(f"Enter marks for subject {i+1}: "))
    total += marks


average = total / subjects
if average >= 80:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 60:
    grade = "C"
elif average >= 50:
    grade = "D"
else:
    grade = "F"
print("\n--- Result ---")
print("Total Marks:", total)
print("Average Marks:", round(average, 2))
print("Grade:", grade)
```
